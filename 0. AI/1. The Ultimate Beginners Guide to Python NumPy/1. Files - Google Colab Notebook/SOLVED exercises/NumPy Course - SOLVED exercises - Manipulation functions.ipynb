### Grouping and splitting arrays

#### `np.concatenate` / `np.stack` / `np.vstack` / `np.hstack`

First use `np.arange` and `reshape` to create a 2D array with values ​​between 0 and 1, and step equal to 0.1, organized as a matrix of 2 rows and 5 columns.

```python
array_0 = np.arange(0, 1, step=0.1).reshape(2, 5)
```

```python
array_0
```

Now create a second array with values ​​between 1 and 2, and step 0.1, in the same format.

```python
array_1 = np.arange(1, 2, step=0.1).reshape(2, 5)
```

```python
array_1
```

Use `np.concatenate` to concatenate the two arrays along the rows, that is, axis 0.

```python
np.concatenate([array_0, array_1], axis=0)
```

👆 Notice that the new array now has 4 rows and 5 columns.

Use `np.vstack` to stack the two arrays vertically, and notice that the result is the same as that of `np.concatenate` with `axis=0`.

```python
np.vstack([array_0, array_1])
```

Use `np.stack` to stack the two arrays on axis 0, but note that in this case, a new dimension is created.

```python
np.stack([array_0, array_1], axis=0)
```

Now repeat the procedure regarding axis 1. In this case, you must use `np.hstack`.

```python
np.concatenate([array_0, array_1], axis=1)
```

```python
np.hstack([array_0, array_1])
```

```python
np.stack([array_0, array_1], axis=1)
```

### Rearranging elements

#### `np.flip`

Invert `array_0`, displaying the result to visualize the change.

```python
np.flip(array_0)
```

👆 The rows and columns have been reversed.

Invert `array_0` only on axis 0.

```python
np.flip(array_0, axis=0)
```

👆 Only the rows were reversed.

Invert `array_0` only on axis 1.

```python
np.flip(array_0, axis=1)
```

👆 Only the columns were reversed.

#### `np.resize`

Resize `array_0` to linearized format with 8 elements.

```python
np.resize(a=array_0, new_shape=8)
```

👆 As the new size is smaller than the previous one, the original array is truncated.

Resize `array_0` to linearized format with 15 elements.

```python
np.resize(a=array_0, new_shape=15)
```

👆 As the new size is larger than the previous one, the original array is repeated up to the new size.

Resize `array_0` to contain 3 rows and 6 columns.

```python
np.resize(a=array_0, new_shape=(3, 6))
```

👆 Remember that the array is first linearized, then truncated or completed to the new size, then the format is adjusted.

#### `np.trim_zeros`

This function only works with 1D arrays.

First create a 1D array with values ​​between 1 and 5, and step = 0.5. Use `np.arange` for this.

```python
array = np.arange(1, 5, step=0.5)
```

```python
array
```

Now use `np.pad` to add 6 values 0 ​​to the left and 4 values 0 ​​to the right of `array`.

```python
array_pad = np.pad(array=array, pad_width=(6, 4), mode='constant', constant_values=0)
```

```python
array_pad
```

Apply `np.trim` to `array_pad` to remove the values 0 ​​from the ends, thus recovering `array`.

```python
array_trim = np.trim_zeros(filt=array_pad)
```

```python
array_trim
```

Repeat the operation using the `trim` parameter to remove only the "front" (`f`) values 0 ​​from `array_pad`.

```python
array_trim = np.trim_zeros(filt=array_pad, trim='f') # b
```

```python
array_trim
```

#### `np.unique`

Let's start by creating a large array of random integers, where there will certainly be repetitions.

First, instantiate a random number generator with `np.random.default_rng`. Set the `seed` to the value 75.

```python
rng = np.random.default_rng(seed=75)
```

Now generate an array with 200 random integer values ​​in the range [50, 75).

```python
array_unique = rng.integers(low=50, high=75, size=200)
```

```python
array_unique
```

With a single call to `np.unique`, get:
- theh unique values
- the indices where each value first appears (`return_index` parameter)
- the indices that can be used to reconstruct the original array from theh unique values ​​(`return_inverse` parameter)
- the count of the unique values ​​(`return_counts` parameter)

```python
unique_values, indices_first, indices_redo, counts = np.unique(ar=array_unique, return_index=True, return_inverse=True, return_counts=True)
```

```python
unique_values
```

```python
indices_first
```

```python
indices_redo
```

```python
counts
```

Does this function work as advertised? 🤔 Use `unique_values` and `indices_redo` to reconstruct the original array.

```python
unique_values[indices_redo]
```
