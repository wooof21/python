#### `np.mod`

The most common way to check whether a number is *even* is to observe whether its *modulus* when divided by 2 is equal to 0.

Use the `np.mod` function to check the modulus of the `array_1` values ​​when divided by 2.

```python
array_mod = np.mod(array_1, 2)
```

```python
array_mod
```

Now use the basic `==` operator to confirm which `array_mod` values ​​are equal to 0.

```python
array_mod == 0
```

This comparison indicates which values ​​are even (`True`).

#### `np.divmod`

Many division operations result in "broken" values. Sometimes we want to represent the result as integer values, in a truncated format, where we discard the decimal part.

Use `np.divmod` to check which is the integer part and the remainder when dividing `array_1` by 4.

```python
array_int, array_rest = np.divmod(array_1, 4)
```

```python
array_int
```

👆 This array represents the rounding by truncation of the results of the operation.

```python
array_rest
```

👆 This is the remainder of the division.

### Trigonometric functions

Calculate the sine, cosine and tangent of some important angles: 30, 45 and 90 degrees. Remember that you need to convert the angle to radians first.

```python
radians_30 = np.radians(30)
```

```python
np.sin(radians_30)
```

👆 The exact value is 0.5.

```python
np.cos(radians_30)
```

```python
np.tan(radians_30)
```

```python
radians_45 = np.radians(45)
```

```python
np.sin(radians_45)
```

```python
np.cos(radians_45)
```

```python
np.tan(radians_45)
```

👆 The exact value is 1.

```python
radians_90 = np.radians(90)
```

```python
np.sin(radians_90)
```

```python
np.cos(radians_90)
```

👆 The exact value is 0.

```python
np.tan(radians_90)
```

👆 The exact value is *infinite*, as $$tan(x) = \frac{sin(x)}{cos(x)}$$ and the cosine of 90 degrees is equal to 0. For rounding reasons, the NumPy returns a "very high" value instead of expressing infinity.

#### `logical_and`, `logical_or`, `logical_xor`, `logical_not`

First generate a boolean mask to check which values ​​in `array_1` and `array_2` are even.

```python
array_1_mask = np.mod(array_1, 2) == 0
```

```python
array_1_mask
```

```python
array_2_mask = np.mod(array_2, 2) == 0
```

```python
array_2_mask
```

Now, with `logical_and`, check where in both arrays the values ​​are even.

```python
np.logical_and(array_1_mask, array_2_mask)
```

With `logical_or` you can check where at least one of the values ​​is even.

```python
np.logical_or(array_1_mask, array_2_mask)
```

You can also check where at least one of the values ​​is odd with `logical_not`, without having to redo the boolean masks.

```python
np.logical_not(array_1_mask, array_2_mask)
```

#### `np.isclose`, `np.allclose`

Now let's generate 2 new arrays with real values ​​from sampling a normal distribution, with a small standard deviation.

```python
array_3 = rng.normal(loc=0, scale=0.0001, size=6).reshape(3, 2)
```

```python
array_3
```

```python
array_4 = rng.normal(loc=0, scale=0.0001, size=6).reshape(3, 2)
```

```python
array_4
```

Check whether the values ​​of `array_3` are close to `array_4` within the 0.001 tolerance margin.

```python
np.isclose(a=array_3, b=array_4, atol=0.001)
```

Check whether all values ​​meet this condition.

```python
np.allclose(a=array_3, b=array_4, atol=0.001)
```

👆 All values ​​are the same within this tolerance range.

What if the tolerance margin is equal to 0.0001?

```python
np.isclose(a=array_3, b=array_4, atol=0.0001)
```

```python
np.allclose(a=array_3, b=array_4, atol=0.0001)
```

👆 Not all values ​​are equal within this tolerance range.
