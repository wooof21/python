In this module, we will work with a 3-D array, which represents the performance of running athletes.

In the first dimension (depth), we will represent 3 different evaluation days. In the second dimension (rows), each of the 5 athletes. In the third dimension (columns), each of the 4 laps of the race track. Values ​​will be given in seconds.

This data will be simulated with the code below, but don't worry about understanding it for now, the details of these functions will be presented in the next classes.

````python
rng = np.random.default_rng(seed=1)
````

````python
rng
````

````python
array = rng.normal(loc=50, scale=3, size=60).reshape(3, 5, 4).round(3)
````

````python
array
````

````python
array.shape
````

````python
3 * 5 * 4
````

````python
array.size
````

👆 Notice from the `shape` that there are 3 days, 5 athletes and 4 laps.

Given the definition of this array presented above, you can understand the very first row:
```
[51,037, 52,465, 50,991, 46,091]
```
as the performance of the first athlete, on the first day, in the 4 laps of the track.

The seventh row:
```
[41,867, 44,333, 49,476, 48,733]
```
is the performance of the second athlete (second row of that "bunch") on the second day (second "bunch"), in the 4 laps of the track.

### Indexing

Given the array above, use indexing to retrieve the performance of the third athlete on the second day.

Remember that in Python, indexing starts at 0!

````python
array
````

````python
array[1, 2]
````

Now recover the performance of all athletes in the last lap of the first day. Use negative indexing to access the last lap.

````python
array[0, :, -1]
````

Now recover the performance of the second and fourth athletes, on the third day, in the second lap.

````python
array[2, [1, 3], 1]
````

Using negative indexing, recover the performance of all athletes on all days, on the penultimate lap.

````python
array[:, :, -2]
````

Create a boolean mask to represent only the results less than 46. Then use this mask to filter the results from the array that meet this condition.

````python
mask = array < 46
````

````python
mask
````

````python
array[mask]
````

### Slicing

Use slicing to recover the fourth athlete's performance in the first two laps, on all days.

````python
array[:, 3, 0:2]
````

Now recover the performance of all athletes in the first two days, from the second lap onwards.

````python
array[0:2, :, 1:]
````

Use negative and step notation to retrieve the performances of the last two athletes, on the second day, in the laps odd-indexed.

````python
array[1, -2:, 1::2]
````

### Assigning

Suppose the second athlete was disqualified in the third day's race, and you must demonstrate this in the array by changing all the values ​​of his laps to the number 99.

Use the assigning technique to make this correction, and display the modified array.

````python
array[2, 1] = 99
````

````python
array
````
