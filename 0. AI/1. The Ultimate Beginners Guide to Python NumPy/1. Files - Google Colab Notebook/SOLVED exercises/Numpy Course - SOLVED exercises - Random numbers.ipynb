#### `np.random.default_rng`

Use the `np.random.default_rng` function to instantiate a random number generator. Use the `seed=101` parameter to ensure code reproducibility.

```python
rng = np.random.default_rng(seed=101)
```

#### `rng.random`


Use `rng.random` to generate a random number in the range [0, 1).

```python
rng.random()
```

Now generate 5 random values ​​in this same range.

```python
rng.random(size=5)
```

Create a 2D array with 3 rows and 4 columns, populated with random values.

```python
rng.random(size=(3, 4))
```

#### `rng.integers`

Generate 8 random integers, whose maximum value (excluded) must be 10.

```python
rng.integers(10, size=8)
```

Generate 7 random integers whose minimum value (included) is 3 and maximum value (excluded) is 20.

```python
rng.integers(low=3, high=20, size=7)
```

Generate 3 random integers where the first is in the range [2, 10), the second in the range [7, 15), and the third in the range [14, 25).

```python
rng.integers(low=[2, 7, 14], high=[10, 15, 25])
```

#### `rng.choice`

Note that `rng.integers` can return repeated values. But you can generate random integers without repetition with `rng.choice`.

First, use `np.arange` to generate an array containing values ​​between 5 and 20 (excluded).

```python
array = np.arange(5, 20)
```

```python
array
```

Now, use `rng.choice` to select 7 values ​​from this array, without repetition. You guarantee that there will be no repetition by passing the `replace=False` parameter.

```python
rng.choice(a=array, size=7, replace=False)
```

#### `rng.shuffle`

Use `rng.shuffle` to permutate the created `array` previously. Remember that the permutation occurs *inplace*, that is, the array is modified and the function returns nothing.

```python
rng.shuffle(array)
```

```python
array
```

#### `rng.permutation`

Now permute `array` again, but with the `rng.permutation` function. In this case, the operation returns a copy of the permuted array.

```python
rng.permutation(array)
```

### Distributions

Each distribution has its own parameters, which indicate how the data will be distributed. As small samples of distributions are difficult to interpret based on their values ​​alone, here you will be invited to plot large random samples of some of the distributions shown in class, but changing the parameters freely. Note the differences between the graphs plotted here and those presented in class. Pay attention to the differences in both the format of the graph and the values ​​of the axes.

```python
plt.hist(rng.beta(a=7, b=50, size=100_000), bins=100);
```

```python
plt.hist(rng.binomial(n=7, p=0.32, size=100_000), bins=100);
```

```python
plt.hist(rng.chisquare(df=12, size=100_000), bins=100);
```

```python
plt.hist(rng.exponential(scale=12, size=100_000), bins=100);
```

```python
plt.hist(rng.normal(loc=10, scale=5, size=100_000), bins=100);
```

```python
plt.hist(rng.uniform(low=20, high=25, size=100_000), bins=100);
```
