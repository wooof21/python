```python
import numpy as np
np.__version__
```

```python
import sklearn
from sklearn.datasets import load_iris
sklearn.__version__
```

In this module we will use the `sklearn` library to load a real dataset.

You can find the documentation here: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris

```python
iris_data = load_iris()
```

```python
iris_data
```

This dataset contains measurements of 4 flower structures from 3 plant species.

In this exercise, we will only use the measurements of the 4 structures.

```python
iris_features = iris_data['data']
```

```python
iris_features
```

```python
iris_features.shape
```

👆 There are 150 flowers, and 4 measurements.

#### `np.percentile`

Calculate the 10th, 25th and 50th percentiles of the 4 variables, that is, "along the rows".

```python
np.percentile(a=iris_features, q=10, axis=0)
```

👆 This means that 10% of plants have the first measurement/variable less than or equal to 4.8, the second less than or equal to 2.5, the third less than or equal to 1.4 and the fourth less than or equal to 0.2.

```python
np.percentile(a=iris_features, q=25, axis=0)
```

👆 This means that 25% (1/4) of the plants have the first measurement less than or equal to 5.1, the second less than or equal to 2.8, the third less than or equal to 1.6 and the fourth less than or equal to 0.3.

```python
np.percentile(a=iris_features, q=50, axis=0)
```

🤔 Can you interpret this result?

#### `np.quantile`

Repeat the previous exercise but now using the `np.quantile` function. Use the same values ​​for the quantiles, but remember that quantiles are percentiles divided by 100. Note that the results are the same, as expected.

```python
np.quantile(a=iris_features, q=0.1, axis=0)
```

```python
np.quantile(a=iris_features, q=0.25, axis=0)
```

```python
np.quantile(a=iris_features, q=0.5, axis=0)
```

#### `np.median`, `np.mean`

Calculate the median and mean of the 4 variables.

```python
np.median(a=iris_features, axis=0)
```

👆 The median is equal to the 50th percentile and the 0.5 quantile.

```python
np.mean(a=iris_features, axis=0)
```

#### `np.std`, `np.var`

Calculate the standard deviation and variance of the 4 variables.

```python
iris_std = np.std(a=iris_features, axis=0)
```

```python
iris_std
```

```python
iris_var = np.var(a=iris_features, axis=0)
```

```python
iris_var
```

Use `np.allclose` to confirm that the variance is equal to the squared standard deviation (`np.power`).

```python
np.allclose(np.power(iris_std, 2), iris_var)
```

#### `np.cov`

Determine the covariance between the 4 variables. Tip: you will need to use the array in its transposed form.

```python
np.cov(m=iris_features.T)
```

#### `np.corrcoef`

Determine the Pearson correlation coefficient between the 4 variables. Again, you will need to use the transposed form of the array.

```python
np.corrcoef(x=iris_features.T)
```

#### `np.histogram`

Determine the value count and intervals to construct a histogram of the variable indexed 2, with 5 intervals in total.

```python
hist, bin_edges = np.histogram(a=iris_features[:, 2], bins=5)
```

```python
hist, bin_edges
```

👆 Note that the majority of plants (50 in total) are in the lower range of the intervals, with values ​​between 1 and 2.18.

#### `np.bincount`

Use `np.bincount` to determine the frequency of each flower in unit intervals, in relation to the variable indexed 2. You will need to enter the values ​​as integers (`astype` method).

```python
np.bincount(iris_features[:, 2].astype(int))
```

👆 There are no flowers in range 0 to 1, but 50 flowers in range 1 to 2.
