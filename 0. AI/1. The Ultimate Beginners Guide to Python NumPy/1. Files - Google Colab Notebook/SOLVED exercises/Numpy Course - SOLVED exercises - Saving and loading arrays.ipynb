```python
import numpy as np
np.__version__
```

Start by generating two arrays:
- the first through sampling from a normal distribution with mean 10 and standard deviation 2, in the 5x3 format.
- the second through sampling from a uniform distribution with a minimum of 25 and a maximum of 50, in the 3x4x2 format.

To ensure reproducibility, use seed equal to 10.

```python
rng = np.random.default_rng(seed=10)
```

```python
array_1 = rng.normal(loc=10, scale=2, size=5*3).reshape(5, 3)
```

```python
array_1
```

```python
array_2 = rng.uniform(low=25, high=50, size=3*4*2).reshape(3, 4, 2)
```

```python
array_2
```

#### `np.save`, `np.load`

Save `array_1` in a file called `array_1_file`.

```python
np.save(file='array_1_file', arr=array_1)
```

Reload the array from the file.

```python
array_1_loaded = np.load(file='array_1_file.npy')
```

```python
array_1_loaded
```

Check that the two arrays are equal with `np.array_equal`.

```python
np.array_equal(array_1, array_1_loaded)
```

#### `np.savez`

Save both arrays in a single `npz` file.

```python
np.savez('arrays_savez', array_1, array_2)
```

Reload the file and restore both arrays.

```python
arrays_loaded = np.load('arrays_savez.npz')
```

```python
arrays_loaded
```

```python
array_1_loaded = arrays_loaded['arr_0']
```

```python
array_1_loaded
```

```python
array_2_loaded = arrays_loaded['arr_1']
```

```python
array_2_loaded
```

#### `np.savez_compressed`

Save both arrays in a single compressed file, then reload.

```python
np.savez_compressed('arrays_compacted', array_1, array_2)
```

```python
arrays_compacted_loaded = np.load('arrays_compacted.npz')
```

```python
arrays_compacted_loaded
```

#### `np.savetxt`, `np.loadtxt`

Save `array_1` in `csv` format.

```python
np.savetxt(fname='array_csv.csv', X=array_1)
```

Use the Linux command below to view how the data was saved to the file.

```python
!head array_csv.csv
```

Reload this array from the file.

```python
array_1_loaded = np.loadtxt(fname='array_csv.csv')
```

```python
array_1_loaded
```

Use `np.array_equal` to verify that even when saved as `csv`, the reloaded array is equal to the original array.

```python
np.array_equal(array_1, array_1_loaded)
```
