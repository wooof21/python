### Matrix and vector products

The following array represents the grades from 5 students in 4 tests applied along a semester.

```python
rng = np.random.default_rng(seed=5)
```

```python
scores = np.clip( # limits the grade in the interval [0, 10]
    a=np.round( # rounds the grades to 1 decimal point
        a=rng.normal(loc=7, scale=2, size=20), # samples numbers from a normal distribution
        decimals=1
        ),
    a_min=0,
    a_max=10
).reshape(5, 4) # reshapes to the desired shape
```

```python
scores
```

👆 5 students (5 rows), 4 grades (4 columns).

To generate the average, the teacher decided to apply different weights to each test, according to the array below.

```python
weights = np.array([[0.2],
                    [0.3],
                    [0.4],
                    [0.1]])
```

```python
weights
```

#### `np.dot`

Use `np.dot` to calculate the product between matrices `scores` and `weights`.

```python
np.dot(a=scores, b=weights)
```

Note that the result represents the **weighted average** of the students.

#### `np.vdot`

Use indexing to get the notes for student indexed 2 in a new variable.

```python
scores_2 = scores[2, :]
```

```python
scores_2
```

Now use `np.vdot` to determine the average grade for student 2. Remeber that, since this function accepts only vectors, you must linearize `weights`, because it is formatted as a matrix. You can use either methods `ravel` or `flatten`, and you can apply the method to `weights` directly from within the `np.vdot` function.

```python
weights.ravel()
```

```python
np.vdot(scores_2, weights.ravel())
```

Note that this is the same result obtained for student 2 in the exercise using `np.dot`.

####  `np.inner`

Redo the same calculation but now using the `np.inner` function.

```python
np.inner(scores_2, weights.ravel())
```

Remember that, for two vectors, the `inner` product is equal to the vector product (`dot` or `vdot`).

#### `np.matmul`

Once more redo the calculation, now using `np.matmul`.

```python
np.matmul(scores_2, weights.ravel())
```

The "matrix" multiplication between two vector does also correspond to the vector product between them.

#### `np.outer`

This function is useful when we want to obtain a matrix from the multiplication of two vectors, where the values are obtained from the multiplication of all values in the vectors.

Given the vectors below:

```python
array_0 = np.array([2, 3, 7])
array_1 = np.array([0.7, 0.2, 0.3, 0.4])
```

Determine the matrix that is obtained by the calculation mentioned above.

```python
np.outer(a=array_0, b=array_1)
```

Observe that the matrix's shape reflects the length of the vectors: 3x4.

#### `np.linalg.det`

The determinant of a matrix is used to calculate the inverse of the matrix.

Calculate the determinant of the following matrix:

```python
array_2 = rng.normal(loc=5, scale=2, size=16).reshape(4, 4)
```

```python
array_2
```

```python
np.linalg.det(array_2)
```

#### `np.linalg.inv`

But you can also obtain the inverse directly from the fuction `np.linalg.inv`. Do this for `array_2`.

```python
array_2_inv = np.linalg.inv(array_2)
```

```python
array_2_inv
```

Use `np.matmul` to confirm that the matrix multiplication between a matrix and its inverse results in the identity matrix, where the values in the diagonal equal 1, and all others are 0.

```python
array_mul = np.matmul(array_2, array_2_inv)
```

```python
array_mul
```

These values are difficult to interpret because of the rounding operations performed by NumPy.

Let's instantiate an identity matrix with the same shape as `array_mul`. This is done with the function `np.identity`.

```python
array_identity = np.identity(n=4)
```

```python
array_identity
```

Finally, use `np.allclose` to verify that indeed the values in `array_mul` are virtually equal to `array_identity`.

```python
np.allclose(a=array_mul, b=array_identity)
```
