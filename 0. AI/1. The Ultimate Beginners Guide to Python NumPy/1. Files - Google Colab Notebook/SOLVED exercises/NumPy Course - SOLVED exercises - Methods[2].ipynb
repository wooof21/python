We will use the same array as in the previous exercises, which represents the performance of 5 athletes in 4 laps of a running track, on 3 different days.

```python
rng = np.random.default_rng(seed=1)
```

```python
array = rng.normal(loc=50, scale=3, size=60).reshape(3, 5, 4).round(3)
```

```python
array
```

```python
array.shape
```

👆 Notice from the `shape` that there are 3 days, 5 athletes and 4 laps.

#### `max` / `argmax`

Remember that, as the values ​​represent the time, in seconds, that each athlete took to complete a lap, the higher this value, the worse the athlete's performance.

Use `max` to find the worst daily performance (axis 0).

```python
array.max(axis=0)
```

👆 The result is an array with 5 rows (corresponding to the 5 athletes) and 4 columns (corresponding to the 4 laps). Each value is the maximum time over the 3 days for that athlete and that lap.

Repeat the exercise with `argmax`.

```python
array.argmax(axis=0)
```

👆 This indicates that, for the first athlete and the first lap, (first row and first column of the visualization) the worst performance was on day 0; for the last athlete and the last lap, the worst performance was on day 1.

Now use `max` and `argmax` to find the worst performance per athlete (axis 1), on each day and on each lap.

```python
array.max(axis=1)
```

👆 This is the worst performance among all athletes, in the 3 days (rows) and in the 4 laps (columns).

```python
array.argmax(axis=1)
```

👆 On the first day (first row), in the 4 laps (columns), the worst athletes were those with index 1, 0, 0 and 3, respectively.

Also find the worst overall performance.

```python
array.max()
```

#### `min`/`argmin`

Repeat the exercise, now with the `mix`/`argmix` methods, to find the best performances.

Interpret the results in a similar way to that done for the worst performers.

#### `ptp`

Find the difference between the worst and best performance, on each day (axis 0).

```python
array.ptp(axis=0)
```

👆 In the case of athlete 0 (first row), his performance fluctuated 4.719 seconds in the first lap, 4.515 seconds in the second lap, and so on.

Find the difference between the worst and best performance, in each lap (axis 2).

```python
array.ptp(axis=2)
```

👆 In the case of lap indexed 1 (second column), the difference in performance between the athletes on the first day was 4.327 seconds, 7.609 seconds on the second day and 1.805 seconds on the third.

#### `clip`

"Clip" this array so that no value is less than 43 or greater than 55. Display the array to view the changes.

```python
array_clip = array.clip(min=43, max=55)
```

```python
array_clip
```

Confirm with `min` and `max` that the change was made as desired.

```python
array_clip.min()
```

```python
array_clip.max()
```

#### `sum`

Use `sum` to view the total time spent by each athlete in each race, over the 3 days. In other words, you must add the time on axis 2, corresponding to the laps.

```python
array_sum_axis_2 = array.sum(axis=2)
```

```python
array_sum_axis_2
```

👆 The array represents the 3 days (3 rows) and the 5 athletes (5 columns).

Now, from this array, add up the performances of the 5 athletes over the 3 days (axis 0).

```python
array_sum_axis_2_axis_0 = array_sum_axis_2.sum(axis=0)
```

```python
array_sum_axis_2_axis_0
```

👆 The array represents the total sum of the performance of the 5 athletes.

#### `cumsum`

Use `cumsum` to find the accumulated time of each athlete, on each day, at the end of each lap (axis 2).

```python
array.cumsum(axis=2)
```

👆 On the second day, the third athlete (row 2) accumulated times of 50.641, 101.293, 157.647 and 204.311.

#### `all` / `any`

First, create a boolean mask (array with `True`/`False` values) to check which values ​​in `array` are greater than 50 seconds.

```python
array_mask = array > 50
```

```python
array_mask
```

Now, use `all` to check if any athlete took more than 50 seconds to complete a lap (axis 2), on any given day.

```python
array_mask.all(axis=2)
```

👆 This indicates that, on the first day (first row), the athlete with index 2 (third column) took more than 50 seconds in all laps.

Use `any` to repeat the analysis, but considering the time spent on any lap.

```python
array_mask.any(axis=2)
```

👆 On the first day, all athletes had at least one lap that took more than 50 seconds; on the second day, only the athlete indexed 1 did not perform worse than 50 seconds.
