#### `np.diag_indices` / `np.tril_indices` / `np.triu_indices`

These functions apply to square matrices.

Start by generating a 2D array of random integers, sampled in the range [0, 50), in 6x2 format.

```python
array = ...
```

```python
array
```

This is a very common way to represent points on a cartesian plane; each row represents a point, the first column is the x coordinate and the second is the y coordinate. In our example, there are 6 points.

With the help of the `matplotlib` library, we will visualize these points.

```python
plt.scatter(array[:, 0], array[:, 1]);
```

Sometimes we want to calculate the distance, pairwise, between all points in a database. To do this, we will use a function from the `sklearn` library.

```python
dist = pairwise_distances(array, array)
```

```python
dist
```

```python
dist.shape
```

Note that the distance matrix has a 6x6 format. This is because it represents the distance of the 6 points from each other. The order of the points is the same as the array: the first row/column represents the first point (first row of `array`), the second row/column represents the second point, and so on.

Also notice that the values ​​on the main diagonal are all equal to 0. This is because this diagonal represents the distance from each point to itself, which is obviously zero.

In this case, as we have few points, it is easy to visually check this property of the distance matrix. But suppose you have a huge matrix, corresponding to thousands of points. How can this verification be carried out?

This happens because, obviously, the distance between points 1 and 2 is equal to the distance between points 2 and 1.

Therefore, to save memory, it is common for distance matrices to have the values ​​in the lower triangle set to zero, as it contains redundant values.
